```python
def is_prime(n: int) -> bool:
    if n <= 1:  # 小于等于1的数不是质数
        return False
    if n <= 3:  # 2和3是质数
        return True
    if n % 2 == 0 or n % 3 == 0:  # 能被2或3整除的不是质数
        return False
    
    # 只需要检查到 sqrt(n)
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True
```

```python
print(is_prime(2))   # True
print(is_prime(4))   # False
print(is_prime(17))  # True
print(is_prime(20))  # False
```
